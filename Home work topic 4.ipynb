```python
# Исходные данные имеют следующую структуру:
# *1. перечень всех документов*
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# *2. перечень полок, на которых хранятся документы (если документ есть в documents, он обязательно должен быть и в directories)*
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def doc_chek(number_doc, documents):
    """
    Проверка наличия документа в списке документов
    Стоило бы проверить по буквам совпадение но я не стал так как результат верный.
    """
    for line in documents:
        result = True
        if number_doc in line['number']: 
            result = True
            break
        else: result = False
    return result

def directories_chek(number_doc, directories):
    """
    Проверка наличия документа на полке.
    """
    result = False
    for key, value in directories.items():
        for mylist in value:
            if result != True:
                if number_doc == mylist:
                    result = True
                else: result = False    
    return result    

def p():
    """
    Проверка и вывод владельца документа или вывод уведомления об отсутствии.
    """
    print('Для отображения информации о владельце документа введите номер документа ')
    number_doc = input()
    for line in documents:
        if doc_chek(number_doc, documents) == True:
            if number_doc in line['number']:
                print('Владелец документа ', line['name'])
                print('\n')
        else:
            print('Владелец документа: владелец не найден')
            break
    return 
    

    
def s():
    """
    Поиск документа на полке и вывод уведомления об отсутствии
    """
    number_doc = input('Для поиска документа на полке введите номер документа ')
    for key, value in directories.items():
        if directories_chek(number_doc, directories) == True:
            for mylist in value:
                if number_doc in mylist:
                    print('Документ хранится на полке ',key)
                    print('\n')
        else:
            print('Документ не найден\n')
            break
    return    

def main():
    """
    Поиск владельца документа и поиск документа на полке.
    """
    while True:
        print('Введите команду из списка: ')
        print('Для отображения информации о владельце документа нажмите - p ')
        print('Для того что бы узнать на какой полке храниться документ нажмите - s ')
        print('Для завершения программы введите q\n')
        command_request = input()
        if command_request == 'q':
            print('Работа завершена ')
            return True
        elif command_request == 'p':
            p()
        elif command_request == 's':
            s()
main()
```
